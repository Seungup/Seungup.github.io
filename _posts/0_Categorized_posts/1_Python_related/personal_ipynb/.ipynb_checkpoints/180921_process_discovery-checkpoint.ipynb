```python
import networkx as nx
import itertools
import matplotlib.pyplot as plt

"""
- process log들은 모두 다음을 따라야 한다
    - 같은 프로세스 모델이라면, source, sink는 모두 같아야 함
    - source, sink는 무조건 하나씩만 존재함. 
    - 항상 xor, and 만 존재함. or 같은 것은 없음
    
    - 지금의 예에서 잘 안되는데, 오히려, 모든 로그에 존재하는 액티비티들을 각각 점으로 잡고, 그 점과 점 사이의 구간을 찾아주는 방식으로 하는게 더 좋을 것 같음. 
        - 안됨, 이렇게 하면 and로 묶이는 것까지 다 들어옴. 
    - 그냥 recursive하게 잘 짜야함. 
    
- 추가로 진행해야 하는 부분
    - ## 현재는 process_instance_lst를 입력받고, 거기서 프로세스 모델을 구성하여 리턴 
    - 향후, 프로세스 로그(dataframe)로부터 처리하는 식으로 변경하는 것이 필요함. 
    - 또한, df로부터 process gen time dist도 함께 고려하여 진행하는 것이 필요함. 
    - xor의 경우 각 activity에 대한 각 선택에 대한 확률값도 고려되어야 함
    - BPMN으로 변환하여 진행하는 것도 고려하는 것이 필요함 
    
- 처음에는 그냥 알파 알고리즘으로 해결하려고 했는데, local dependency에서 문제가 발생함. 아래와 같은 예에 대해서 제대로 된 프로세스 모델을 만들어주지 못함. 
    ['a', 'c', 'b', 'd'], 
    ['a', 'b', 'c', 'd'], 
    ['a', 'e', 'd']
- 이제 퍼펙트하게 되지만, 다만 해당 프로세스 모델로 인해 발생하는 모든 로그에 대한 경우의 수가 로그에 존재해야 함 
    - 예를 들어 다음처럼 xor(bb, cc, dd) ==> xor(ff, gg)일 경우 모두 보여져야 함. 
    - 그렇지 않을 경우, 
    ['aa', 'bb', 'ff', 'zz'], 
    ['aa', 'cc', 'ff', 'zz'], 
    ['aa', 'dd', 'ff', 'zz'], 
    ['aa', 'bb', 'gg', 'zz'], 
    ['aa', 'cc', 'gg', 'zz'], 
    ['aa', 'dd', 'gg', 'zz'], 
"""

def make_process_model_from_log(process_instance_lst):
    def make_DiGraph_lst(process_instance_lst):
        ## 입력받은 각각의 프로세스 인스턴스로부터 DiGraph를 생성 
        rG = nx.DiGraph()
        ## add nodes
        nodes = list(set(itertools.chain.from_iterable(process_instance_lst)))
        rG.add_nodes_from(nodes)
        ## add edges 
        for p in process_instance_lst:
            edge_lst = [(p[i], p[i+1]) for i in range(0, len(p)-1)]
            rG.add_edges_from(edge_lst)
        return list(nx.weakly_connected_component_subgraphs(rG))
    
    def find_source_sink(inputDiG):
        ## inputDiG의 source, sink 모두 찾아서 리턴함, 없을 경우 None을 리턴
        sources = [k for k, v in inputDiG.in_degree if v==0]
        sinks = [k for k, v in inputDiG.out_degree if v==0]
        return sources, sinks

    def make_process_from_g(inputDiG):
        ## 입력받은 DiGraph로부터 프로세스 모델을 도출함. 
        ## recursive하게 풀어서 subprocess에 대해서도 진행됨 
        if len(inputDiG.nodes())==0:
            return []
        elif len(inputDiG.nodes())==1:
            return [('act', list(inputDiG.nodes())[0])]
        else:
            subg_lst = list(nx.weakly_connected_component_subgraphs(inputDiG))
            if len(subg_lst)==1:## 모두 연결되어 있는 상황이므로, source, sink를 찾아서 진행해줘야 함 
                newG = subg_lst[0].copy()
                sources, sinks = find_source_sink(newG)
                newG.remove_nodes_from(sources+sinks)
                if len(sources)==0:
                    if len(sinks)==0:
                        return [('and', [('act', n) for n in newG.nodes()])]
                    elif len(sinks)==1:
                        return make_process_from_g(newG)+[('act', sinks[0])]
                    elif len(sinks)>1:
                        sinks = [('act', s) for s in sinks]
                        return make_process_from_g(newG)+[('xor', sinks)]
                elif len(sources)==1:
                    if len(sinks)==0:
                        return [('act', sources[0])]+make_process_from_g(newG)
                    elif len(sinks)==1:
                        return [('act', sources[0])]+make_process_from_g(newG)+[('act', sinks[0])]
                    elif len(sinks)>1:
                        sinks = [('act', s) for s in sinks]
                        return [('act', sources[0])]+make_process_from_g(newG)+[('xor', sinks)]
                elif len(sources)>1:
                    sources = [('act', s) for s in sources]
                    if len(sinks)==0:
                        return [('xor', sources)]+make_process_from_g(newG)
                    elif len(sinks)==1:
                        return [('xor', sources)]+make_process_from_g(newG)+[('act', sinks[0])]
                    elif len(sinks)>1:
                        sinks = [('act', s) for s in sinks]
                        return [('xor', sources)]+make_process_from_g(newG)+[('xor', sinks)]
            elif len(subg_lst)>1:## 복수의 sub-network가 존재하는 상태 
                xor_lst = []
                for subg in subg_lst:
                    #xor_lst+=make_process_from_g(subg)
                    elem = make_process_from_g(subg)
                    if len(elem)==1:## 사이즈가 1일 경우는, 해당 subprocess가 하나의 액티비티만 포함할 때 
                        xor_lst.append(elem[0]) 
                    else:## 사이즈가 1보다 길 경우는, 해당 subprocess가 복합 프로세스일 경우 
                        xor_lst.append( ('seq', elem) ) 
                return [('xor', xor_lst)]
            else:
                return []
            #####
    ## main code 
    rlst = []
    for DiG in make_DiGraph_lst(process_instance_lst):
        rlst.append( ('seq', make_process_from_g(DiG)) )
    return rlst
process_instance_lst = [
    ['a', 'b', 'z'], 
    ['a', 'c', 'z'], 
    ['a', 'd', 'f', 'z'], 
    ['a', 'f', 'd', 'z'], 
    ['a', 'g', 'h', 'z'], 
    ['a', 'i', 'h', 'z'], 
    ['aa', 'bb', 'ff', 'zz'], 
    ['aa', 'cc', 'ff', 'zz'], 
    ['aa', 'dd', 'ff', 'zz'], 
    ['aa', 'bb', 'gg', 'zz'], 
    ['aa', 'cc', 'gg', 'zz'], 
    ['aa', 'dd', 'gg', 'zz'], 
    ['aaa', 'bbb', 'ccc']
]
temp_l = make_process_model_from_log(process_instance_lst)

for i, p_m in enumerate(temp_l):
    print(i, p_m)
print("=="*20)
"""
- 이제 이걸 graph로 변환하고, 
"""
```

```python
def to_BPMN(input_p_m):
    def expand_subprocess(subprocess_data, inputG, source, sink):
        inputG.remove_edges_from([(source, sink)])
        for x in subprocess_data:
            if x[0]=='seq':
                inputG.add_nodes_from([ (x[1], {'type':'activity'}) ])
                inputG.add_edges_from([ (source, x[1]), (x[1], sink)])
            else:
                this_gatewary_num = len([n for n in inputG.nodes(data=True) if n[1]['type']=='gateway'])
                gateway_name = f"{x[0]}_{this_gatewary_num:0>2d}"
                rG.add_nodes_from([ (gateway_name, {'type':'gateway', 'subprocess':x[1]}) ] )
                ordered_nodes.append(gateway_name)
    rG = nx.DiGraph()
    ## add node
    ordered_nodes = []
    for x in input_p_m:
        if x[0]=='act':
            rG.add_nodes_from([ (x[1], {'type':'activity'}) ])
            ordered_nodes.append(x[1])
        else:
            num_label = len([n for n in rG.nodes(data=True) if n[1]['type']==x[0]])
            node_label = f"{x[0]}_{num_label:0>2d}"
            rG.add_nodes_from([(node_label, {'type':'xor',
                                             'subprocess':x[1]
                                            })])
            ordered_nodes.append(node_label)
    ## add edge 
    rG.add_edges_from([(ordered_nodes[i], ordered_nodes[i+1]) for i in range(0, len(ordered_nodes)-1)])
    print("=="*20)
    print(rG.nodes(data=True))
    print(rG.edges())
    return rG
```

```python
import itertools
import functools
import networkx as nx 

def return_unique_activities(input_log):
    uniq_act = itertools.chain.from_iterable(input_log)
    return list(set(uniq_act))

def return_all_direct_succession(input_log):
    rlst = []
    for instance in input_log:
        rlst += [(instance[i], instance[i+1]) for i in range(0, len(instance)-1)]
    return list(set(rlst))

def causality_matrix(input_log):
    uniq_activity = return_unique_activities(input_log)
    dir_successions = return_all_direct_succession(input_log)
    
    causality_matrix = { key1: {} for key1 in uniq_activity}
    for a1 in uniq_activity:
        for a2 in uniq_activity:
            if (a1, a2) in dir_successions and (a2, a1) in dir_successions:
                causality_matrix[a1][a2]="||"
            elif (a1, a2) in dir_successions and (a2, a1) not in dir_successions:
                causality_matrix[a1][a2]="->"
            elif (a1, a2) not in dir_successions and (a2, a1) in dir_successions:
                causality_matrix[a1][a2]="<-"
            else:
                causality_matrix[a1][a2]="#"
    return causality_matrix

def return_transitions(input_log):
    T_L = return_unique_activities(input_log)
    T_I = list(set( [instance[0] for instance in input_log]))
    T_O = list(set( [instance[-1] for instance in input_log]))
    return (T_L, T_I, T_O)

def find_subsets(lst):
    # A, B 내부 집단의 a끼리, b끼리는 관계가 없고, 모든 a, 모든 b 간에는 direct succession이 있는 것들
    subsets = [list(itertools.combinations(lst, i)) for i in range(1, len(lst)+1)]
    subsets = itertools.chain.from_iterable(subsets)
    return list(subsets)

def return_X_L(input_log):
    T_L, T_I, T_O = return_transitions(input_log)
    c_matrix = causality_matrix(input_log)
    A_s = find_subsets(T_L)[:-1]# remove last item
    AB_s = [(A, B) for A in A_s for B in find_subsets( list(set(T_L) - set(A)) )]
    def check_direction(A, B):
        return all( c_matrix[a][b]=="->" for a in A for b in B)
    def check_no_relation(A, B):
        return all( c_matrix[a][b]=="#" for a in A for b in B)
    AB_s = filter(lambda AB: check_direction(AB[0], AB[1]), AB_s)
    AB_s = filter(lambda AB: check_no_relation(AB[0], AB[0]), AB_s)
    AB_s = filter(lambda AB: check_no_relation(AB[1], AB[1]), AB_s)
    X_L = list(AB_s)
    return X_L

def return_Y_L(input_log):
    # make it maximal
    X_L = return_X_L(input_log)
    Y_L = X_L.copy()
    remove_lst =[]
    for i in range(0, len(X_L)-1):
        for j in range(i+1, len(X_L)):
            if set(X_L[i][0]).issubset(set(X_L[j][0])):
                if set(X_L[i][1]).issubset(set(X_L[j][1])):
                    if X_L[i] not in remove_lst:
                        remove_lst.append(X_L[i])
    for rem_elem in remove_lst:
        Y_L.remove(rem_elem)
    return Y_L

def return_P_L(input_log):
    T_L, T_I, T_O = return_transitions(input_log)
    Y_L = return_Y_L(input_log)
    P_L = [("P"+str(i+1), {"From":Y_L[i][0], "To":Y_L[i][1]}) for i in range(0, len(Y_L))]
    P_L.insert(0, ("source", {"From":(), "To":[elem for elem in T_I]}))
    P_L.append(("sink", {"From":[elem for elem in T_O], "To":()}))
    return P_L

def return_F_L(input_log):
    P_L = return_P_L(input_log)
    F_L = [(a, p_name) for p_name, p_attr in P_L for a in p_attr["From"]]
    F_L +=[(p_name, a) for p_name, p_attr in P_L for a in p_attr["To"]]
    return F_L

def alpha_algorithm(input_log):
    T_L, T_I, T_O = return_transitions(input_log)
    P_L = return_P_L(input_log)
    F_L = return_F_L(input_log)
    return (P_L, T_L, F_L)

def make_process_model_from_log(input_log):
    rDiG = nx.DiGraph()
    P_L, T_L, F_L = alpha_algorithm(input_log)
    rDiG.add_nodes_from([ (p[0], {'type':'place'}) for p in P_L])
    rDiG.add_nodes_from([ (p[0], {'type':'transition'}) for p in T_L])
    rDiG.add_edges_from([ e for e in F_L])
    return rDiG

input_log = [
    ['a', 'c', 'b', 'd'], 
    ['a', 'b', 'c', 'd'], 
    ['a', 'e', 'd']
]

p_m = make_process_model_from_log(input_log)
print(p_m.nodes())
print(p_m.edges())

plt.figure(figsize=(12, 6))
nx.draw_networkx(p_m, pos=nx.spring_layout(p_m))
```
