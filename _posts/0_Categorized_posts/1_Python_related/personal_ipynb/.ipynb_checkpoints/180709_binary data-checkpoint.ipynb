```python
## what is base64
"""
text와 binary data의 차이 
파일을 텍스트로 읽을때, binary로 읽을때의 차이.

샘플 텍스트 파일을 만들고, 이를 그냥 읽을때와 rb로 읽을때의 차이를 구분해서 보여주면 되지 않을까? 
엄밀하게 따지면 모든 파일은 binary 데이터다. 
binary data: 
"""

import base64

msg = "안녕하세요".encode()
print(msg)

encoded_msg = base64.b64encode(msg)
print( encoded_msg)
print( base64.b64decode(encoded_msg))
print( base64.b64decode(encoded_msg).decode('utf-8'))
#base64.b64encode(msg)
```

```python
msg = """내가 임마 느그 스장이랑 모굑탕도 가고\n"""
f = open('test_text.txt', 'w')
f.write(msg)
f.close()

f = open("test_text.txt", 'r')
print("read text:\n{}\n".format(f.read()))
f.close()

f = open("test_text.txt", 'rb')
bin_text = f.read()
print("read binary:\n{}\n".format(bin_text))

base64.b64decode(base64.b64encode(msg.encode()))
```

```python
"""
write binary ==> read binary 
write text ==> read binary 
"""
msg = """내가 임마 느그 스장이랑 모굑탕도 가고\n"""

f = open('test_text.txt', 'w')
f.write(msg), f.close()
f = open('test_text.bin', 'wb')
f.write(msg.encode()), f.close()

f_txt = open("test_text.txt", 'r')
f_bin = open('test_text.bin', 'rb')
```

```python
import requests
import matplotlib.pyplot as plt

url = "https://amueller.github.io/word_cloud/_images/sphx_glr_colored_003.png"
response = requests.get(url)
print("binary file sample: {}".format(response.content[:20]))

## binary file을 만들어줍니다. 
f = open("write_bin_img.png", 'wb')
f.write(response.content), f.close()

## 저장된 binary file을 다시 이미지로 읽어서, 잘 저장되어 있는지를 확인합니다. 
img_np_array = plt.imread('write_bin_img.png')
f = plt.figure()
plt.imshow(img_np_array)
plt.savefig("../../assets/images/markdown_img/180709_bin_img.svg")
```
