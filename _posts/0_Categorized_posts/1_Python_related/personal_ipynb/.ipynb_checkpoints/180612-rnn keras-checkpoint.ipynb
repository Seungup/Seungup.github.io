```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import math

from keras.models import Sequential
from keras.layers import Dense, LSTM

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error

data_url = '/Users/frhyme/Downloads/international-airline-passengers.csv'
df = pd.read_csv(data_url)[:-1]
df.columns=['m', 'p']

plt.figure(figsize=(12, 2))
plt.plot(df['p'])
plt.show()

np.random.seed(7) # fix the random seet for reproducibility

data = df['p'].values.astype('float32')
scaler = MinMaxScaler()
data = scaler.fit_transform(data.reshape(-1, 1))
train, test = data[:len(data)//3*2], data[len(data)//3*2:]

def create_dataset(dataset, look_back=1):
    # look_back만큼을 새로운 feature로 만들어서 넣어줌.
    # 예를 들어, look_back이 3이라면, 3일전, 2일전, 1일 전 이라는 세 가지 feature가 만들어짐. 
    # 이렇게 보면 그냥 regression과 큰 차이 없는 것으로 보임. 
    dataX, dataY = [], []
    for i in range(len(dataset)-look_back-1):
        a = dataset[i:(i+look_back), 0]
        dataX.append(a)
        dataY.append(dataset[i + look_back, 0])
    return np.array(dataX), np.array(dataY)

look_back = 1
trainX, trainY = create_dataset(train, look_back)
testX, testY = create_dataset(test, look_back)
"""
reshape to (samples, time steps, features)
"""
trainX = np.reshape(trainX, (trainX.shape[0], 1, trainX.shape[1]))
testX = np.reshape(testX, (testX.shape[0], 1, testX.shape[1]))

## make rnn model 
model = Sequential([
    LSTM(4, input_shape=(1, look_back)), 
    Dense(1)
])
model.compile(loss='mean_squared_error', optimizer='adam')
"""
- cnn 때는 batch_size가 컸는데, 여기서는 일단 1
- 그리고 데이터에 비해서 rnn의 epoch이 매우 큰 편이라는 것을 알 수 이따. 
"""
model.fit(trainX, trainY, epochs=30, batch_size=1, verbose=0)
print("complete")
```
