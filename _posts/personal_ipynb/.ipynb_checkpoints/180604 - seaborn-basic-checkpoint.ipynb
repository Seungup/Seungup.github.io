```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

lst = np.random.normal(0, 10, 200)

"""
그냥 hist와의 차이라면, 비율로 나온다는 것 과, kernel density function을 그려준다는 것. 
"""

plt.figure(figsize=(12, 4))
sns.distplot(lst, bins=30, rug=True)
#plt.savefig('../../assets/images/markdown_img/180604_not_applied_seaborn.svg')
plt.show()
```

```python
"""
plt.scatter와 유사하지만, 위, 아래에 hist plot을 함께 그려줘서, 데이터의 분포를 대략적으로 알아볼 수 있어서 좋다.
"""
sample_size = 1000

X = np.random.normal(0, 1, sample_size)
Y = X + np.random.normal(0, 1, sample_size)

sns.jointplot(X, Y, alpha=0.5, marker='o', color='red')
plt.show()
```
