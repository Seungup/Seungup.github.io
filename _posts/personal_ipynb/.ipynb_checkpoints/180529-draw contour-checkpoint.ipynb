```python
import numpy as np
import matplotlib.pyplot as plt
xlist = np.linspace(-3.0, 3.0, 10)
ylist = np.linspace(-3.0, 3.0, 10)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)
plt.figure()
cp = plt.contourf(X, Y, Z)
plt.colorbar(cp)
plt.title('Filled Contours Plot')
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')
plt.show()
```

```python
xlist = np.linspace(-3.0, 3.0, 10)
ylist = np.linspace(-3.0, 3.0, 10)

XX, YY = np.meshgrid(np.linspace(-3.0, 3.0, 1000),
                     np.linspace(-3.0, 3.0, 1000)
                    )
print("XX.shape: {}".format(XX.shape))
print("YY.shape: {}".format(YY.shape))
ZZ = np.sqrt(XX**2 + YY**2 )

plt.figure(figsize=(12, 5))
"""levels에 구간을 넣어줘서 등고선 표시 위치를 정할 수 있습니다. 
"""
cp = plt.contour(XX, YY, ZZ, 
                 levels = np.linspace(ZZ.reshape(-1, 1).min(), ZZ.reshape(-1, 1).max(), 50)
                )
plt.colorbar(cp)
plt.xlabel("X")
plt.show()
```
