```python
import networkx as nx
import numpy as np 
import matplotlib.pyplot as plt
from itertools import combinations
```

```python
g = nx.Graph()

g.add_nodes_from([("n{:0>2d}".format(i), {'weight':np.random.normal(0, 1)}) for i in range(0, 20)])
g.add_edges_from([e for e in combinations(g.nodes(), 2) if np.random.normal(0, 1) < -1])
plt.figure(figsize=(15, 8))
pos = nx.spring_layout(g)
node_color = [n[1]['weight'] for n in g.nodes(data=True)]
nx.draw_networkx_nodes(g, pos=pos, nodesize=500, 
                       node_color = node_color, 
                       cmap=plt.cm.Reds, )

plt.axis('off')
sm = plt.cm.ScalarMappable(cmap=plt.cm.Reds, norm=[-0.1, 1.0])
sm._A = []
plt.colorbar(sm)
plt.show()
[n[1]['weight'] for n in g.nodes(data=True)]

```
