```python
import pandas as pd
import datetime as dt 
import matplotlib.pyplot as plt
import seaborn as sns

weeknum_to_dict = {0:'mon', 1:'tue', 2:'wed', 3:'thu', 4:'fri', 5:'sat', 6:'sun'}

df = pd.DataFrame(list(map(lambda s: s.split("\t"), input_text.split("\n"))), 
                  columns = ['date', 'time', 'cost']
                 )
# date preprocessing
df['date_time'] = df['date']+df['time']
df['date_time'] = df['date_time'].apply(lambda x: dt.datetime.strptime(x, '%Y. %m. %d.%H:%M'))
df = df[['date_time', 'cost']]
df['cost'] = df['cost'].astype('float64')
# weekday preprocessing

df['weekday'] = df['date_time'].apply(lambda d: str(d.weekday())+"_"+weeknum_to_dict[d.weekday()])
df = df.set_index('date_time')
# 안간 날짜도 카운트해서 넣을 것 
#print(df.head())

date_index = pd.date_range(df.index.min().date(), df.index.max().date())
df_all_date = pd.Series([0 for i in range(0, len(date_index))], index=date_index)
timestamp = pd.Series([[] for i in range(0, len(date_index))], index=date_index)
for row in df.iterrows():
    df_all_date[row[0].date()]+=row[1]['cost']
    timestamp[row[0].date()].append(row[0])
df_all_date = pd.DataFrame(df_all_date, columns=['cost'])

df_all_date['weekday'] = [str(d.weekday())+"_"+ weeknum_to_dict[d.weekday()] for d in df_all_date.index]
df_all_date['timestamp'] = timestamp
df_all_date['study_time'] = df_all_date['timestamp'].apply(
    lambda ts: 0 if len(ts)==0 else (max(ts)-min(ts)+dt.timedelta(2/24, 0, 0)).seconds/3600
)
df_all_date = df_all_date[2:]# 3월부터 고려함
#df_all_date['date'] = [d.date() for d in df_all_date.index]
df_all_date.head(10)
```

```python
"""
- 공부시간과 cost는 비슷한 그래프를 따르니까 그냥 cost로만 해줘도 될것 같고 
- 폰트를 바꾸고, 
- marker를 어떤 icon으로 바꿀 수 있는지를 좀 보면 좋을 것 같다. 
"""
plt.figure(figsize=(8,4 ))
df_all_date.resample('w')['cost'].sum().plot(marker='o', ax=plt.gca())
plt.yticks([])
plt.xlim(df_all_date.index.min()-7, df_all_date.index.max()+7)
plt.show()
```

```python
#요일별 커피 사용금액 변화
plt.figure(figsize=(8,4 ))
df_all_date.groupby('weekday')['cost'].sum().plot(marker='o', linestyle='--')
plt.show()
```

```python
"""
3월 초에는 11일 연속으로 가놓고, 그다음에 바로 5일 연속으로 쳐놀음 깔깔깔 
"""
df_cost_grp_week = df_all_date.reset_index().pivot('index', 'weekday', 'cost').fillna(0).resample('w').sum()
df_cost_grp_week.index = ["{}, {}w".format(ind.strftime("%b"), ind.day//7 + 1) for ind in df_cost_grp_week.index]
df_cost_grp_week.columns = [col[2:].upper() for col in df_cost_grp_week.columns]
plt.figure(figsize=(10, 10))
sns.heatmap(df_cost_grp_week, 
            linewidths=(5), 
            cmap=plt.cm.Reds, 
            cbar=False, 
            annot=True, fmt=".0f", 
            mask = df_cost_grp_week.applymap(lambda x: True if x<=1000 else False)
)
plt.tick_params(labelsize=12,)
plt.gca().xaxis.tick_top() 
plt.xlabel('')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
from matplotlib.cbook import get_sample_data

def main():
    x = np.linspace(0, 10, 20)
    y = np.cos(x)
    image_path = get_sample_data('/Users/frhyme/Downloads/orange-coffee-512.png')
    fig, ax = plt.subplots()
    imscatter(x, y, image_path, zoom=0.1, ax=ax)
    ax.plot(x, y)
    plt.show()

def imscatter(x, y, image, ax=None, zoom=1):
    if ax is None:
        ax = plt.gca()
    try:
        image = plt.imread(image)
    except TypeError:
        # Likely already an array...
        pass
    im = OffsetImage(image, zoom=zoom)
    x, y = np.atleast_1d(x, y)
    artists = []
    for x0, y0 in zip(x, y):
        ab = AnnotationBbox(im, (x0, y0), xycoords='data', frameon=False)
        artists.append(ax.add_artist(ab))
    ax.update_datalim(np.column_stack([x, y]))
    ax.autoscale()
    return artists

main()
```
