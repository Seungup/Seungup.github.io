```python
import numpy as np

"""일부러 
"""

test_arr = np.array([i for i in range(0, 2*3*5*7)]).reshape(2, 3,5,7)
print(test_arr)
print("---")
print("shape: {}".format(test_arr.shape))
```

```python

print(np.min(test_arr, axis=None))
print("---")

for axis_v in range(0, len(test_arr.shape)):
    print(np.min(test_arr, axis=axis_v))
    print("---")
```
